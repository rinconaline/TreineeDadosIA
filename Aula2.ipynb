```python
import pandas as pd
```

```python
df = pd.read_csv("archive/hantavirus_detection_dataset.csv")
df
```

```python
df = df.drop(columns=['Patient_ID'])
df = df.drop(columns=['Immunological_Risk'])
df = df.drop(columns=['Disease_Onset_Date'])
df = df.drop(columns=['Sample_Date'])
df = df.drop(columns=['Gender'])
df = df.drop(columns=['Age'])
df = df.drop(columns=['Hantavirus_Type'])
df
```

```python
df.info()
```

```python
df.memory_usage(deep=True)
```

```python
df['Symptoms'].unique()
```

```python
df['Symptoms'] = df['Symptoms'].map({'Fever': 1, 'Headache': 2, 'Muscle_Pain': 3, 'Nausea': 4, 'Vomiting': 5, 'Diarrhea': 6, 'Abdominal_Pain': 7, 'Cough': 8, 'Shortness_of_Breath': 9, 'Chills': 10, 'Dizziness': 11, 'Rash': 12, 'Bleeding': 13})
```

```python
df['Renal_Risk'] = df['Renal_Risk'].map({'Low': 1, 'Medium': 2, 'High': 3})
```

```python
df['Disease_Severity'] = df['Disease_Severity'].map({'Mild': 1, 'Moderate': 2, 'Severe': 3})
```

```python
df['Region'] = df['Region'].map({'North': 1, 'South': 2, 'East': 3, 'West': 4, 'Central': 5})
```

```python
df['Exposure_Type'] = df['Exposure_Type'].map({'Occupational': 1, 'Recreational': 2, 'Residential': 3, 'Unknown': 4})
```

```python
df.describe()
```

```python
df.to_csv("archive/hantavirus_detection_dataset_preprocessed.csv", index=False)
```
