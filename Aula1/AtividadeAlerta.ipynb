```python
valores_brutos: List[int] = [10,50,150,20,300]

def verificar_valores(lista: List[int]) -> List[int]:
    alertas = []

    for valor in lista:
        if valor > 100:
            alertas.append(valor)
    if not alertas:
        print("Sistema Seguro")
    else:
        print(f"⚠️Revisar valores: {alertas}")
```
